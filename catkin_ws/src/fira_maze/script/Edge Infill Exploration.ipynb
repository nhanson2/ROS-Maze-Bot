```python
import numpy as np
import cv2

# Will use matplotlib instead of cv2 to show the image
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Load an color image in grayscale
img = cv2.imread('/home/nathaniel/ROS-Maze-Bot/catkin_ws/src/fira_maze/maps/part_1_complete__map.pgm',-1)

# show image
canny = cv2.Canny(img, 100,100)
contours, _ = cv2.findContours(canny, 2,2)
length = len(contours)
for i in range(len(contours)):
    cv2.drawContours(img, contours, i, (0,255,0), 2, cv2.LINE_8)
plt.figure(figsize = (80,80))
plt.imshow(img)
```
